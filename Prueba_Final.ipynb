# Problema 1:
Un biólogo quiere simular la población de peces en un lago para estudiar su comportamiento a lo largo del tiempo. La simulación debe seguir estas reglas:
1.	Cada año, la población de peces se duplica.
2.	Cada año, un porcentaje fijo de la población de peces es pescado.
3.	Cada año, un porcentaje fijo de la población de peces muere por causas naturales.

Escribe un programa que simule la población de peces durante 10 años, dado un número inicial de peces un porcentaje de pesca del 23% y un porcentaje de mortalidad natural el 9%. El programa debe imprimir la población de peces al final de cada año. Usted desconoce la población inicial de peces, solamente el biólogo tiene esa información.


```python
def simulacion_pesca():
    poblacion_inicial = int(input("Ingrese la población inicial de peces: "))


    porcentaje_inicial = 0.23
    porcentaje_mortalidad = 0.09
    poblacion_actual = poblacion_inicial

    for año in range(1, 11):
        poblacion_actual *= 2
        poblacion_actual -= poblacion_actual * porcentaje_inicial
        poblacion_actual -= poblacion_actual * porcentaje_mortalidad
        print(f"Año {año}: Población de peces = {int(poblacion_actual)}")

simulacion_pesca()
```

#Problema 2:
Una empresa quiere analizar sus datos mensuales de ventas para identificar tendencias y desempeño. Los datos están almacenados en un archivo CSV que contiene las siguientes columnas: Fecha de venta, producto vendido, cantidad vendida y precio unitario. El programa debe permitir:
1.	Cargar los datos desde el archivo CSV.
2.	Calcular el total de ventas mensuales.
3.	Calcular el producto más vendido por mes.
4.	Graficar la evolución de las ventas a lo largo de los meses.

Escribe un programa que permita cargar, analizar y visualizar estos datos de ventas mensuales.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#1 Cargar los datos desde el archivo CSV
df= pd.read_excel("/content/drive/MyDrive/Python/Taller Nine Hub_0624/Bases de datos/Base de datos_Prueba.xlsx")
df.info()
df.head()
```

```python
#2 Calcular el total de ventas mensuales
df['Total Vendido'] = df['Cantidad Vendida'] * df['Precio Unitario']
df['Mes'] = df['Fecha de Venta'].dt.to_period('M')
total_ventas = df.groupby('Mes')['Total Vendido'].sum()
print(total_ventas)
```
